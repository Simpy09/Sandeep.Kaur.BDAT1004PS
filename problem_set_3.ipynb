```python
# Step 1: Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset from the given URL
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|')

# Step 3: Assign it to a variable called users
print(users.head()) # just to verify if the data was imported correctly
```

```python
# step 4: Group the users by occupation and calculate the mean age for each group
mean_age_by_occupation = users.groupby('occupation')['age'].mean()

# Print the result
print(mean_age_by_occupation)
```

```python
# step 5: Group the users by occupation and gender
users_by_occupation_gender = users.groupby(['occupation', 'gender'])['gender'].count()

# Group the users by occupation
users_by_occupation = users.groupby('occupation')['gender'].count()

# Calculate the male ratio for each occupation
male_ratio_by_occupation = users_by_occupation_gender.loc[:, 'M'] / users_by_occupation * 100

# Sort the male ratio in descending order
male_ratio_by_occupation_sorted = male_ratio_by_occupation.sort_values(ascending=False)

# Print the result
print(male_ratio_by_occupation_sorted)
```

```python
# step 6: Group the users by occupation and calculate the minimum and maximum age for each group
min_max_age_by_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

# Print the result
print(min_max_age_by_occupation)
```

```python
# step 7: Group the users by occupation and gender, and calculate the mean age for each group
mean_age_by_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()

# Print the result
print(mean_age_by_occupation_gender)
```

```python
# step 8: Group the users by occupation and gender
users_by_occupation_gender = users.groupby(['occupation', 'gender'])['gender'].count()

# Group the users by occupation
users_by_occupation = users.groupby('occupation')['gender'].count()

# Calculate the percentage of women and men for each occupation
percentage_by_occupation_gender = users_by_occupation_gender / users_by_occupation * 100

# Print the result
print(percentage_by_occupation_gender)
```

```python
# Step 1: Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Import the dataset
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Step 3: Clean the data
data = data.dropna() # Remove missing values
data = data[(data['Year'] >= 1867) & (data['Year'] <= 2014)] # Filter years

# Step 4: Create the line plot
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages')
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces')
plt.xlabel('Year')
plt.ylabel('Number per capita')
plt.title('Number of marriages and divorces per capita in the U.S. (1867-2014)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the dataset
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Step 2: Clean and filter the data
data = data.dropna(subset=['Marriages_per_1000', 'Divorces_per_1000']) # Remove missing values
years = [1900, 1950, 2000]
data = data[data['Year'].isin(years)] # Filter years

# Step 3: Create the bar chart
marriages = data['Marriages_per_1000']
divorces = data['Divorces_per_1000']

plt.bar(years, marriages, align='center', alpha=0.5, label='Marriages')
plt.bar(years, divorces, align='center', alpha=0.5, label='Divorces')
plt.xlabel('Year')
plt.ylabel('Number per capita')
plt.title('Number of marriages and divorces per capita in the U.S. (1900, 1950, 2000)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the data from the CSV file
data = pd.read_csv('actor_kill_counts.csv')

# Step 2: Group the data by actor and sum the kills for each actor
grouped_data = data.groupby(['Actor'])['Count'].sum()

# Step 3: Sort the actors by their kill count
sorted_data = grouped_data.sort_values(ascending=True)

# Step 4: Create a horizontal bar chart
plt.barh(sorted_data.index, sorted_data)

# Step 5: Label each bar with the corresponding actor's name
for index, value in enumerate(sorted_data):
    plt.text(value, index, str(value))

# Step 6: Label the axes
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data
data = pd.read_csv('roman-emperor-reigns.csv')

# Step 2: Group by cause of death and count the number of emperors
counts = data.groupby('Cause_of_Death').count()['Emperor']

# Step 3: Create the pie chart
plt.pie(counts, labels=counts.index, autopct='%1.0f%%')
plt.title('Fraction of Assassinated Roman Emperors')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
data = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')

# Filter data by year
data = data[(data['Year'] >= 2000) & (data['Year'] <= 2009)]

# Create a scatter plot with color by year
plt.scatter(data['Total Arcade Revenue (billions)'], data['Computer Science Doctorates Awarded (US)'], c=data['Year'])

# Add labels and title
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.title('Relationship between Arcade Revenue and CS Doctorates (2000-2009)')
plt.colorbar(label='Year')
plt.show()
```
